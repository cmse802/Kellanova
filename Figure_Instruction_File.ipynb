# Data Preperation

It is important to note that Tableau is not a data cleaning software and has minimal power to edit the data when loaded in to the software. Tableau can filter and adjust things as necessary, but the main objective of Tableau is to be able to drag and drop the columns of data accordingly to create a visualization (more on this down below.)

Consider using some sort of data manipulation software or code, for example python and the pandas library to load in the data set, alter and clean it, and the re-export it as a new .csv file or similar in order to have a better experience working with Tableau.

# Loading A Dataset

Tableau is a power data visualization tool and thus can use many different types of data files that are loaded in when first opening the app. Find a simple data set to use for the next step of this process. Heres some of the more common data file types that Tableau can load from. Note there's plenty more that are supported, and a quick google search can determine if its an acceptable file type.

   1. Excel spreadsheets : .xls, .xlsx
   2. Comma Separated values: .csv
   3. Text files: .txt
   4. JSON files: .json

Once a data set has been selected, you will need to navigate to its location on your computer. The specific location of this file is not important, so store the data file wherever is convenient, organized or necessary for you.

   To load this data set and navigate to it, on the Tableau home page look to the left side of the screen. There you will see different sections labeled "Search for Data", " To a File", "To a Server" and "Saved Data Sources"
   
   Locate the "To a File" section, and find the "More..." item at the bottom of that section and click it. This will open up the file explorer where you can then navigate to the data file you've selected. 
   
   Select the data file and click open. This then brings us to a window where you can look at the data fields, columns, rows etc.
   
   To start working on plotting itself, locate the glowing orange 'Sheet 1' at the bottom left corner of the screen and click it. This will bring us to the actual Tableau workspace.

# Creating A Visualization 

Once your data is imported, you will then be able to manipulate your sheet by creating calculated fields, applying filters, and sorting data as needed. The core process of building a visualization in tableau is the drag and drop interface, where you select and place your data dimensions (qualitative and quantitative data) onto the rows and columns of your sheet to shape your visualization. These are called 'pills.' This automatically generates your graph based on your inputs and you have a plethora of options in terms of chart types, data manipulations, colors, and labels to enhance interactivity and understanding. To improve upon your plot further you can introduce additional filters to hone in on specific aspects of your data. 

Additionally, you can create a dashboard to combine multiple visualizations and have a comprehensive view of your analysis. This dashboard is created by clicking the top tab labeled 'Dashboard' and click 'new dashboard.' From here you can drag and drop in your created Tableau worksheets. In this dashboard, the different plots can interact with one another and act as a filter. There's limitless possibilities using a combination of worksheets and dashboards. and it's up to the user to find what they like to showcase the data.

# Exporting a Visualization

To export a visualization in tableau it is quite trivial, what you’ll want to do is right click on the visualization (sheet tab at the bottom) you want to export then you can choose the title you want to save it as and where you want the file to be exported to. You can also go to file and choose to export the file as a couple different data types such as PDF, PowerPoint, or as an image, etc.
